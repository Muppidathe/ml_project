```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Muppidathe/ml_project/refs/heads/master/UniversalBank.csv')
df
```

```python
df=df.drop(columns=['ID','ZIP Code'])
df
```

```python
df.info()
```

```python
x=df.drop(columns=['Personal Loan'])
y=df['Personal Loan']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,stratify=y,random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
model=SVC()
param_grid = {
    'model__C': [0.1, 1, 10],
    'model__kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'model__gamma': ['scale', 'auto'],
    'model__degree': [2, 3, 4],          # only affects 'poly' kernel
    'model__coef0': [0.0, 0.1, 0.5]      # for 'poly' and 'sigmoid'
}
pipe=Pipeline(steps=[('scaler',MinMaxScaler()),('model',model)])
grid=GridSearchCV(estimator=pipe,param_grid=param_grid,cv=5,scoring='accuracy',verbose=2,)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
y_pred_train=grid.predict(x_train)
accuracy_train=accuracy_score(y_train,y_pred_train)
precision_train=precision_score(y_train,y_pred_train)
recall_train=recall_score(y_train,y_pred_train)
print('train accuracy ',accuracy_train)
print('train precision ',precision_train)
print('train recall ',recall_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
y_pred_test=grid.predict(x_test)
accuracy_test=accuracy_score(y_test,y_pred_test)
precision_test=precision_score(y_test,y_pred_test)
recall_test=recall_score(y_test,y_pred_test)
print('test accuracy ',accuracy_test)
print('test precision ',precision_test)
print('test recall ',recall_test)
```
