```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Muppidathe/ml_project/refs/heads/master/logistic_regression_dataset.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
encoderL=LabelEncoder()
encoderO=OrdinalEncoder()
new_df=tfidf.fit_transform(df['Review']).toarray()
new_df=pd.DataFrame(new_df,columns=tfidf.get_feature_names_out())
new_df['label_encod_category']=encoderL.fit_transform(df['Category'])
new_df['label_encod_Rating']=encoderO.fit_transform(df[['Rating']])
new_df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new_df.drop(columns=['label_encod_Rating']),new_df['label_encod_Rating'],test_size=.2,random_state=42)
x_train,y_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.get_params
```

```python
import numpy as np
review=tfidf.transform(['very good book to learn Data science']).toarray()
category=encoderL.transform(['Books'])
x=np.hstack((review[0],category))
model.predict([x])
```

# trying ***BOW***

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text
documents = ["I love cats", "Cats love me"]

# Initialize BoW model
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(documents)

# Convert to array
print("Vocabulary:", vectorizer.get_feature_names_out())
print("BoW Matrix:\n", bow_matrix.toarray())
```
