```python
import pandas as pd
```

```python
df = pd.DataFrame({ 'Age': [25, 30, 22, 35, 40, 28, 42, 36], 'Income': [40, 60, 25, 70, 65, 45, 80, 75] })
df
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=2,random_state=0)
model.fit_predict(df)
```

```python
df = pd.DataFrame({ 'Age': [25, 30, 22, 35, 40, 28, 42, 36], 'Income': [40, 60, 25, 70, 65, 45, 80, 75], 'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male'], 'Occupation': ['Engineer', 'Doctor', 'Artist', 'Lawyer', 'Engineer', 'Artist', 'Doctor', 'Lawyer'] })
df
```

```python
from kmodes.kprototypes import KPrototypes
model=KPrototypes(n_clusters=2,random_state=0,verbose=2,n_jobs=-1,)
model.fit_predict(df,categorical=[2,3])
```
