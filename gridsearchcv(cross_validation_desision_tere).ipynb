```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Muppidathe/ml_project/refs/heads/master/loan_default_risk_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df=df.dropna(axis=0)
```

```python
sum(df.duplicated())
```

```python
df.isnull().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['Loan_Default_Risk']),df['Loan_Default_Risk'],train_size=.8,random_state=42)
x_train.shape,y_train.shape
```

```python
model=DecisionTreeClassifier(max_depth=5,random_state=42)
model.fit(x_train,y_train)
```

```python
model.get_depth()
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import precision_score,accuracy_score,recall_score
precision=precision_score(y_test,y_pred)
accuracy=accuracy_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
print(f'Accuracy {accuracy} \nprecision {precision} \nrecall {recall}')
```

```python
from sklearn.model_selection import GridSearchCV
model1=DecisionTreeClassifier(random_state=42)
grid=GridSearchCV(model1,param_grid={'max_depth':[1,2,3,4,5,6,7,8,9,None],'criterion':["gini", "entropy", "log_loss"]},cv=3,scoring='accuracy',)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
from sklearn.metrics import precision_score,accuracy_score,recall_score
precision=precision_score(y_test,y_pred)
accuracy=accuracy_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
print(f'Accuracy {accuracy} \nprecision {precision} \nrecall {recall}')
```
